```python
model = AutoModelForCausalLM.from_pretrained("microsoft/Phi-3-mini-4k-instruct", device_map="cpu", torch_dtype="auto", trust_remote_code=True)
```

```python
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
from transformers import pipeline
```

```python
generator = pipeline("text-generation", model=model, tokenizer=tokenizer, return_full_text=False, max_new_tokens=500, do_sample=False)
```

```python
messages=[{"role":"user", "content":"Create a funny joke about chickens."}]
output = generator(messages)
print(output[0]["generated_text"])
```

```python
prompt = "Create a funny joke about chickens."
tokens = generator.tokenizer.tokenize(prompt)
print(tokens)
```
